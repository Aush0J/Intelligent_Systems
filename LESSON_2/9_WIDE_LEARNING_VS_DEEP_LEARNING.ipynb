```python
# import libraries
import torch
import torch.nn as nn
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
from ucimlrepo import fetch_ucirepo 
from sklearn.preprocessing import MinMaxScaler
# fetch dataset 
iris = fetch_ucirepo(id=53) 
  
# data (as pandas dataframes) 
# Initialize the MinMaxScaler
scaler = MinMaxScaler()
X = iris.data.features 
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
y = iris.data.targets
y["class"].unique()
y = y.copy()
y["class"] = y["class"].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
# convert to a pytorch tensor
data = torch.tensor(X.values).float()
labels = torch.tensor(y.values)#.float()
labels = labels.view(-1, 1)
labels=labels.squeeze().long()
```

# Construct and sanity-check the model

# Create a function that trains the model

```python
# a function to train the model

def train(model):

    # define the loss function and optimizer
    lossfunc = nn.CrossEntropyLoss()
    optimizer = torch.optim.SGD(model.parameters(),lr=.01)

    # loop over epochs
    for i in range(epochs):

        # forward pass
        pred = model(data)

        # compute loss
        loss = lossfunc(pred,labels)

        # backprop
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()



    # final forward pass to get accuracy
    predictions = model(data)
    predlabels = torch.argmax(predictions,axis=1)
    acc = 100*torch.mean((predlabels == labels).float())

    # total number of trainable parameters in the model
    nParams = sum(p.numel() for p in model.parameters() if p.requires_grad)

    # function outputs
    return acc,nParams
```

# Now for the experiment!

```python
# define the model parameters
num_layers = range(1,11)         # number of hidden layers
num_units  = np.arange(4,101,3) # units per hidden layer

# initialize output matrices
accuracies  = np.zeros((len(num_units),len(num_layers)))
totalparams = np.zeros((len(num_units),len(num_layers)))

# number of training epochs
epochs = 1000

# Perform Experiment
for unitidx in range(len(num_units)):
    for layeridx in range(len(num_layers)):

        # create a fresh model instance
        net = ANN(num_units[unitidx],num_layers[layeridx])

        # run the model and store the results
        acc,nParams = train(net)
        accuracies[unitidx,layeridx] = acc

        # store the total number of parameters in the model
        totalparams[unitidx,layeridx] = nParams
```
